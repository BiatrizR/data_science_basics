```python
url = 'https://api.covid19api.com/dayone/country/brazil'
#requisão da url
resp = r.get(url)
```

```python
# Separar/filtrar os dados chaves que vamos utilizar:
final_data = []
for obs in raw_data:
    final_data.append([obs['Confirmed'], obs['Deaths'], obs['Recovered'], obs['Active'], obs['Date']])
    
final_data 
```

```python

# formatando a data, pegandoapenas até o caracter de dia/mes/ano:
for i in range(1, len(final_data)):
    final_data[i][DATA] = final_data[i][DATA][:10]

final_data
```

 ### Criando gráficos
   - API Quickchart

```python
def get_datasets(y, labels):
    if type(y[0]) == list:
        datasets = []
        for i in range(len(y)):
            datasets.append({
                'label': labels[i],
                'data': y[i]
           })
        return datasets
    else:
        return [
           {
               'label':labels[0],
               'data': y
           }
       ]
       
```

```python
def set_title(title = ''):
    if title != '':
        display = 'true'
    else:
        display = 'false'   
    return {
        'title': title, 
        'display': display
    }
```

```python
#criar dicinário que representará o gráfico
#valor x, y, as legendas do grárido, o tipo(nesse caso de barras),tituloO(caso tenha) 

def create_chart(x, y, labels, kind='bar', title=''):

    datasets = get_datasets(y, labels)
    options = set_title(title)
    
    chart = {
        'type': kind,
        'data': {
            'labels': x,
            'datasets':datasets
        },
        'options': options
    }
    
    return chart
```

```python
#fazer a requisição na api utilizando a estrutura do dicionário acima
```

```python
#requisição
def get_api_chart(chart):
    url_base = 'https://quickchart.io/chart'
    resp = r.get(f'{url_base}?c={str(chart)}')
    #retornar um valor binário para armazenar a imagem
    return resp.content
```

```python
#salvar a imagem
def save_image(path, content):
    with open(path, 'wb') as image:
        image.write(content)
```

```python
y_data_1 = []
for obs in final_data[1::30]:
    y_data_1.append(obs[CONFIRMADOS])
    
y_data_2 = []
for obs in final_data[1::30]:
    y_data_2.append(obs[RECUPERADOS])

labels = ['Confirmados', 'Recuperados']

x =[]
for obs in final_data[1::30]:
    x.append(obs[DATA].strftime('%d/%m/%Y'))
    
chart = create_chart(x, [y_data_1, y_data_2], labels, title='Gráfico confirmados vs recuperados')
chart_content = get_api_chart(chart)
save_image('meu-primeiro-grafico.png', chart_content)
display_image('meu-primeiro-grafico.png')    
```

```python
def get_api_qrcode(link):
    text = quote (link) #parsing link para url
    url_base = 'https://quickchart.io/qr'
    resp = r.get(f'{url_base}?text={text}')
    return resp.content
```
